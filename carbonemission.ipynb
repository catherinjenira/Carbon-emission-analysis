```python
import plotly.express as px

fig = px.line(
	data_frame=emission,
	x="Country name",
	y=1990,
	markers=True,
	title="World carbon emission (1990)"
)
fig.show()
```

```python
# Prepare df2 for plotting
df2 = emission.melt(
	id_vars=['Country name'],
	value_vars=[year for year in range(1990, 2012)],
	var_name='Year',
	value_name='value'
)
df2 = df2[df2['value'] != '..']  # Remove missing values
df2['value'] = pd.to_numeric(df2['value'], errors='coerce')
df2 = df2.dropna(subset=['value'])

fig = px.choropleth(
	data_frame=df2[df2['Country name'] != 'World'],
	locationmode='country names',
	locations='Country name',
	color='value',
	animation_frame='Year',
	title='CO2 Emission by Countries by Year',
	color_continuous_scale=px.colors.sequential.RdBu_r,
	range_color=(df2['value'].max(), df2['value'].min())
)
fig.show()
```

```python
%pip install raceplotly

from raceplotly.plots import barplot
race_plot = barplot(
	df=df2[df2['Country name'] != 'World'],
	item_column='Country name',
	value_column='value',
	time_column='Year',
	top_entries=10
)
race_plot.plot(
	title='Carbon Emission Race by Country (Since 1990 to 2011)',
	value_label='Value',
	item_label='Top 10 Countries'
)
```
